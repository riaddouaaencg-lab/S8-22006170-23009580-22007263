```python
# ── CELLULE 1 : Installations & Imports ──────────────────────

!pip install scikit-learn pandas numpy matplotlib seaborn -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
import matplotlib.patheffects as pe
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, roc_curve, precision_recall_curve,
                             average_precision_score)
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.inspection import permutation_importance
import warnings
warnings.filterwarnings('ignore')

# Palette corporative
DARK_BG    = "#0D1117"
CARD_BG    = "#161B22"
ACCENT_RED = "#E63946"
ACCENT_YLW = "#F4A261"
ACCENT_GRN = "#2EC4B6"
ACCENT_BLU = "#457B9D"
TEXT_MAIN  = "#E6EDF3"
TEXT_SUB   = "#8B949E"

print("✅  Imports réussis — prêt pour l'analyse !")


```

```python
# ── CELLULE 2 : Création du Dataset ──────────────────────────

np.random.seed(42)
n_employes = 1000

data = {
    'ID_Employe': range(1, n_employes + 1),
    'Departement': np.random.choice(
        ['Achats', 'Finance', 'Ventes', 'RH', 'IT', 'Logistique'],
        n_employes, p=[0.15, 0.15, 0.25, 0.10, 0.15, 0.20]
    ),
    'Anciennete_Annees':        np.random.randint(1, 25, n_employes),
    'Score_Pression_Financiere': np.random.randint(1, 11, n_employes),
    'Satisfaction_Travail':      np.random.randint(1, 11, n_employes),
    'Heures_Supp_Mois':          np.random.randint(0, 50, n_employes),
    'Conges_Non_Pris':           np.random.randint(0, 30, n_employes),
    'Acces_Privilegie':          np.random.choice([0, 1], n_employes, p=[0.8, 0.2])
}
df = pd.DataFrame(data)

# Variable cible — Triangle de Cressey
score_risque = (
    (df['Score_Pression_Financiere'] * 0.4) +
    (df['Acces_Privilegie'] * 15) +
    ((10 - df['Satisfaction_Travail']) * 0.3) +
    (df['Anciennete_Annees'] * 0.2)
)
seuil = np.percentile(score_risque, 95)
df['Fraude_Interne'] = (score_risque >= seuil).astype(int)
df['Score_Risque']   = score_risque

# Encodage département
le = LabelEncoder()
df['Dept_encoded'] = le.fit_transform(df['Departement'])

print(f"📊  Dataset : {len(df)} employés | {df['Fraude_Interne'].sum()} cas de fraude ({df['Fraude_Interne'].mean()*100:.1f} %)")
df.head()
```

```python
# ── CELLULE 3 : Figure 1 — Vue d'ensemble & Triangle de Cressey ──

fig = plt.figure(figsize=(20, 12), facecolor=DARK_BG)
fig.suptitle("PRÉDICTION DE LA FRAUDE INTERNE EN ENTREPRISE",
             fontsize=22, fontweight='bold', color=TEXT_MAIN, y=0.98)
fig.text(0.5, 0.94, "Basé sur le Triangle de la Fraude — ACFE Report 2024",
         ha='center', fontsize=13, color=TEXT_SUB)

gs = gridspec.GridSpec(2, 3, figure=fig, hspace=0.45, wspace=0.35,
                       left=0.06, right=0.97, top=0.91, bottom=0.07)

# ── KPI Cards ────────────────────────────────────────────────
ax_kpi = fig.add_subplot(gs[0, :])
ax_kpi.set_xlim(0, 10); ax_kpi.set_ylim(0, 1); ax_kpi.axis('off')

kpis = [
    ("👥", "Effectif Total",   f"{len(df):,}", ACCENT_BLU),
    ("🚨", "Cas de Fraude",    f"{df['Fraude_Interne'].sum()}", ACCENT_RED),
    ("📈", "Taux de Fraude",   f"{df['Fraude_Interne'].mean()*100:.1f}%", ACCENT_YLW),
    ("🔑", "Accès Privilégié", f"{df['Acces_Privilegie'].mean()*100:.0f}%", ACCENT_GRN),
    ("😟", "Pression Moy.",    f"{df['Score_Pression_Financiere'].mean():.1f}/10", "#C77DFF"),
]
for i, (icon, label, val, col) in enumerate(kpis):
    x = 0.5 + i * 1.85
    rect = FancyBboxPatch((x - 0.82, 0.05), 1.6, 0.88,
                          boxstyle="round,pad=0.05", facecolor=CARD_BG,
                          edgecolor=col, linewidth=2)
    ax_kpi.add_patch(rect)
    ax_kpi.text(x, 0.74, icon,   ha='center', va='center', fontsize=22)
    ax_kpi.text(x, 0.50, val,    ha='center', va='center', fontsize=20,
                fontweight='bold', color=col)
    ax_kpi.text(x, 0.22, label,  ha='center', va='center', fontsize=10,
                color=TEXT_SUB)

# ── Triangle de Cressey ──────────────────────────────────────
ax_tri = fig.add_subplot(gs[1, 0])
ax_tri.set_xlim(0, 10); ax_tri.set_ylim(0, 10); ax_tri.axis('off')
ax_tri.set_facecolor(DARK_BG)
ax_tri.set_title("🔺 Triangle de la Fraude\n(Cressey, 1953)", color=TEXT_MAIN,
                 fontsize=12, fontweight='bold', pad=10)

triangle_pts = np.array([[5, 9], [1, 2], [9, 2], [5, 9]])
ax_tri.fill(triangle_pts[:, 0], triangle_pts[:, 1],
            alpha=0.15, color=ACCENT_RED)
ax_tri.plot(triangle_pts[:, 0], triangle_pts[:, 1],
            color=ACCENT_RED, linewidth=2.5)

nodes = [
    (5, 9.3, "OPPORTUNITÉ",   "Accès Privilégié\nContrôles Faibles", ACCENT_GRN),
    (0.3, 1.2, "PRESSION",    "Stress Financier\nObjectifs Irréalistes", ACCENT_YLW),
    (9.7, 1.2, "RATIONALISATION", "Insatisfaction\nSentiment d'Injustice", "#C77DFF"),
]
for xn, yn, title, sub, col in nodes:
    ax_tri.text(xn, yn, title, ha='center', va='bottom', fontsize=9,
                fontweight='bold', color=col)
    ax_tri.text(xn, yn - 0.9, sub, ha='center', va='top', fontsize=7.5,
                color=TEXT_SUB, linespacing=1.4)

ax_tri.text(5, 5.3, "FRAUDE\nINTERNE", ha='center', va='center',
            fontsize=13, fontweight='bold', color=ACCENT_RED, alpha=0.9)

# ── Fraude par département ───────────────────────────────────
ax_dept = fig.add_subplot(gs[1, 1])
ax_dept.set_facecolor(DARK_BG)
dept_stats = df.groupby('Departement').agg(
    total=('Fraude_Interne', 'count'),
    fraudes=('Fraude_Interne', 'sum')
).assign(taux=lambda x: x['fraudes'] / x['total'] * 100).sort_values('taux', ascending=True)

colors_bar = [ACCENT_RED if v > dept_stats['taux'].mean() else ACCENT_BLU
              for v in dept_stats['taux']]
bars = ax_dept.barh(dept_stats.index, dept_stats['taux'],
                    color=colors_bar, edgecolor='none', height=0.6)
for bar, v in zip(bars, dept_stats['taux']):
    ax_dept.text(v + 0.2, bar.get_y() + bar.get_height()/2,
                 f'{v:.1f}%', va='center', fontsize=9, color=TEXT_MAIN)
ax_dept.axvline(dept_stats['taux'].mean(), color=ACCENT_YLW,
                linestyle='--', linewidth=1.5, alpha=0.8)
ax_dept.set_title("Taux de Fraude par Département", color=TEXT_MAIN,
                  fontsize=11, fontweight='bold')
ax_dept.set_xlabel("Taux de fraude (%)", color=TEXT_SUB, fontsize=9)
ax_dept.tick_params(colors=TEXT_SUB, labelsize=9)
for spine in ax_dept.spines.values(): spine.set_visible(False)
ax_dept.set_facecolor(DARK_BG)
ax_dept.xaxis.label.set_color(TEXT_SUB)

# ── Distribution du score de risque ─────────────────────────
ax_dist = fig.add_subplot(gs[1, 2])
ax_dist.set_facecolor(DARK_BG)
non_fraud = df[df['Fraude_Interne'] == 0]['Score_Risque']
fraud     = df[df['Fraude_Interne'] == 1]['Score_Risque']
ax_dist.hist(non_fraud, bins=35, color=ACCENT_BLU, alpha=0.7,
             label='Non-Fraudeur', density=True, edgecolor='none')
ax_dist.hist(fraud, bins=20, color=ACCENT_RED, alpha=0.8,
             label='Fraudeur', density=True, edgecolor='none')
ax_dist.axvline(seuil, color=ACCENT_YLW, linestyle='--',
                linewidth=1.8, label=f'Seuil = {seuil:.1f}')
ax_dist.set_title("Distribution du Score de Risque", color=TEXT_MAIN,
                  fontsize=11, fontweight='bold')
ax_dist.set_xlabel("Score de risque", color=TEXT_SUB, fontsize=9)
ax_dist.set_ylabel("Densité", color=TEXT_SUB, fontsize=9)
ax_dist.legend(fontsize=8, facecolor=CARD_BG, labelcolor=TEXT_MAIN,
               edgecolor='none')
ax_dist.tick_params(colors=TEXT_SUB, labelsize=9)
for spine in ax_dist.spines.values(): spine.set_visible(False)

plt.savefig('fig1_overview.png', dpi=150, bbox_inches='tight',
            facecolor=DARK_BG)
plt.show()
print("✅  Figure 1 enregistrée")

```

```python
# ── CELLULE 5 : Entraînement des modèles ML ──────────────────

features = ['Dept_encoded', 'Anciennete_Annees', 'Score_Pression_Financiere',
            'Satisfaction_Travail', 'Heures_Supp_Mois', 'Conges_Non_Pris',
            'Acces_Privilegie']
feat_labels = ['Département', 'Ancienneté', 'Pression Financière',
               'Satisfaction', 'Heures Supp.', 'Congés Non Pris', 'Accès Privilégié']

X = df[features]; y = df['Fraude_Interne']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)

# Random Forest
rf = RandomForestClassifier(n_estimators=200, max_depth=8, random_state=42,
                             class_weight='balanced', min_samples_leaf=3)
rf.fit(X_train, y_train)
y_prob_rf = rf.predict_proba(X_test)[:, 1]
y_pred_rf = rf.predict(X_test)

# Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=150, learning_rate=0.08,
                                 max_depth=4, random_state=42)
gb.fit(X_train, y_train)
y_prob_gb = gb.predict_proba(X_test)[:, 1]
y_pred_gb = gb.predict(X_test)

# Régression Logistique
lr = LogisticRegression(random_state=42, class_weight='balanced',
                         max_iter=1000, C=0.5)
lr.fit(X_train_s, y_train)
y_prob_lr = lr.predict_proba(X_test_s)[:, 1]
y_pred_lr = lr.predict(X_test_s)

# Cross-validation AUC
cv  = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
auc_rf_cv = cross_val_score(rf, X, y, cv=cv, scoring='roc_auc').mean()
auc_gb_cv = cross_val_score(gb, X, y, cv=cv, scoring='roc_auc').mean()
auc_lr_cv = cross_val_score(lr, scaler.fit_transform(X), y,
                             cv=cv, scoring='roc_auc').mean()

models_perf = {
    'Random Forest':       (y_pred_rf, y_prob_rf, auc_rf_cv, ACCENT_GRN),
    'Gradient Boosting':   (y_pred_gb, y_prob_gb, auc_gb_cv, ACCENT_YLW),
    'Logistic Regression': (y_pred_lr, y_prob_lr, auc_lr_cv, ACCENT_BLU),
}
for name, (pred, prob, auc_cv, _) in models_perf.items():
    auc = roc_auc_score(y_test, prob)
    print(f"  {name:<25}  AUC test = {auc:.4f}  |  AUC CV-5 = {auc_cv:.4f}")

print("\n✅  Modèles entraînés !")

```

```python
# ── CELLULE 6 : Figure 3 — Performances des modèles ─────────

fig = plt.figure(figsize=(20, 12), facecolor=DARK_BG)
fig.suptitle("PERFORMANCES DES MODÈLES DE CLASSIFICATION",
             fontsize=18, fontweight='bold', color=TEXT_MAIN, y=1.01)

gs = gridspec.GridSpec(2, 3, figure=fig, hspace=0.45, wspace=0.38)

# ── Courbes ROC ──────────────────────────────────────────────
ax_roc = fig.add_subplot(gs[0, 0])
ax_roc.set_facecolor(DARK_BG)
ax_roc.plot([0, 1], [0, 1], '--', color=TEXT_SUB, linewidth=1, alpha=0.5)
for name, (pred, prob, auc_cv, col) in models_perf.items():
    fpr, tpr, _ = roc_curve(y_test, prob)
    auc = roc_auc_score(y_test, prob)
    ax_roc.plot(fpr, tpr, color=col, linewidth=2.5, label=f"{name} (AUC={auc:.3f})")
ax_roc.fill_between(*roc_curve(y_test, y_prob_rf)[:2],
                    alpha=0.08, color=ACCENT_GRN)
ax_roc.set_title("Courbes ROC", color=TEXT_MAIN, fontweight='bold')
ax_roc.set_xlabel("Taux Faux Positifs", color=TEXT_SUB)
ax_roc.set_ylabel("Taux Vrais Positifs", color=TEXT_SUB)
ax_roc.legend(facecolor=CARD_BG, labelcolor=TEXT_MAIN,
               edgecolor='none', fontsize=8)
ax_roc.tick_params(colors=TEXT_SUB)
for sp in ax_roc.spines.values(): sp.set_visible(False)

# ── Précision-Rappel ─────────────────────────────────────────
ax_pr = fig.add_subplot(gs[0, 1])
ax_pr.set_facecolor(DARK_BG)
for name, (pred, prob, auc_cv, col) in models_perf.items():
    prec, rec, _ = precision_recall_curve(y_test, prob)
    ap = average_precision_score(y_test, prob)
    ax_pr.plot(rec, prec, color=col, linewidth=2.5, label=f"{name} (AP={ap:.3f})")
baseline = y_test.mean()
ax_pr.axhline(baseline, color=TEXT_SUB, linestyle='--',
              linewidth=1, alpha=0.6, label=f"Baseline ({baseline:.2f})")
ax_pr.set_title("Courbes Précision-Rappel", color=TEXT_MAIN, fontweight='bold')
ax_pr.set_xlabel("Rappel", color=TEXT_SUB)
ax_pr.set_ylabel("Précision", color=TEXT_SUB)
ax_pr.legend(facecolor=CARD_BG, labelcolor=TEXT_MAIN,
              edgecolor='none', fontsize=8)
ax_pr.tick_params(colors=TEXT_SUB)
for sp in ax_pr.spines.values(): sp.set_visible(False)

# ── Matrice de confusion RF ──────────────────────────────────
ax_cm = fig.add_subplot(gs[0, 2])
ax_cm.set_facecolor(DARK_BG)
cm = confusion_matrix(y_test, y_pred_rf)
cmap_cm = sns.light_palette(ACCENT_RED, as_cmap=True)
sns.heatmap(cm, annot=True, fmt='d', ax=ax_cm, cmap=cmap_cm,
            linewidths=2, linecolor=DARK_BG,
            xticklabels=['Prédit Non-Fraude', 'Prédit Fraude'],
            yticklabels=['Réel Non-Fraude', 'Réel Fraude'],
            annot_kws={'size': 16, 'weight': 'bold', 'color': TEXT_MAIN})
ax_cm.set_title("Matrice de Confusion (RF)", color=TEXT_MAIN, fontweight='bold')
ax_cm.tick_params(colors=TEXT_SUB, labelsize=9)

# ── Feature Importance ───────────────────────────────────────
ax_fi = fig.add_subplot(gs[1, 0])
ax_fi.set_facecolor(DARK_BG)
importances = rf.feature_importances_
idx = np.argsort(importances)
colors_fi = [ACCENT_RED if importances[i] == max(importances) else ACCENT_BLU
             for i in idx]
bars = ax_fi.barh([feat_labels[i] for i in idx], importances[idx],
                   color=colors_fi, edgecolor='none', height=0.6)
for bar, v in zip(bars, importances[idx]):
    ax_fi.text(v + 0.002, bar.get_y() + bar.get_height()/2,
               f'{v:.3f}', va='center', fontsize=8.5, color=TEXT_MAIN)
ax_fi.set_title("Importance des Variables (RF)", color=TEXT_MAIN, fontweight='bold')
ax_fi.set_xlabel("Importance (Gini)", color=TEXT_SUB)
ax_fi.tick_params(colors=TEXT_SUB, labelsize=9)
for sp in ax_fi.spines.values(): sp.set_visible(False)

# ── Comparatif AUC (cross-val) ───────────────────────────────
ax_auc = fig.add_subplot(gs[1, 1])
ax_auc.set_facecolor(DARK_BG)
noms   = ['Random Forest', 'Gradient\nBoosting', 'Logistic\nRegression']
aucs   = [auc_rf_cv, auc_gb_cv, auc_lr_cv]
cols_b = [ACCENT_GRN, ACCENT_YLW, ACCENT_BLU]
bars_auc = ax_auc.bar(noms, aucs, color=cols_b, edgecolor='none',
                       width=0.5, alpha=0.9)
for bar, v in zip(bars_auc, aucs):
    ax_auc.text(bar.get_x() + bar.get_width()/2, v + 0.003,
                f'{v:.4f}', ha='center', va='bottom',
                fontsize=11, fontweight='bold', color=TEXT_MAIN)
ax_auc.set_ylim(0.5, 1.05)
ax_auc.axhline(0.5, color=TEXT_SUB, linestyle='--', linewidth=1, alpha=0.5)
ax_auc.set_title("AUC — Validation Croisée (k=5)", color=TEXT_MAIN, fontweight='bold')
ax_auc.set_ylabel("Score AUC", color=TEXT_SUB)
ax_auc.tick_params(colors=TEXT_SUB, labelsize=9)
for sp in ax_auc.spines.values(): sp.set_visible(False)

# ── Rapport de classification (texte) ────────────────────────
ax_rep = fig.add_subplot(gs[1, 2])
ax_rep.set_facecolor(CARD_BG)
ax_rep.axis('off')
report = classification_report(y_test, y_pred_rf, target_names=['Non-Fraude', 'Fraude'])
ax_rep.text(0.05, 0.95, "Rapport — Random Forest", transform=ax_rep.transAxes,
            fontsize=11, fontweight='bold', color=ACCENT_GRN, va='top')
ax_rep.text(0.05, 0.78, report, transform=ax_rep.transAxes,
            fontsize=9.5, color=TEXT_MAIN, va='top', family='monospace',
            linespacing=1.6)
rect = FancyBboxPatch((0.02, 0.02), 0.96, 0.96,
                       boxstyle="round,pad=0.02",
                       facecolor='none', edgecolor=ACCENT_GRN,
                       linewidth=1.5, transform=ax_rep.transAxes)
ax_rep.add_patch(rect)

plt.savefig('fig3_models.png', dpi=150, bbox_inches='tight', facecolor=DARK_BG)
plt.show()
print("✅  Figure 3 enregistrée")

```
